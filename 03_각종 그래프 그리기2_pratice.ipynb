# 4. 파이차트 그리기
## 4.1 파이차트
- 각 범주(Category)가 데이터에서 차지하는 비율을 나타내는데 사용
- `pie(x, labels)` 이용
    - x: 값 (값들을 100을 기준으로 비율을 계산해 크기 설정)
    - labels : 값들의 label -> 무슨값을 표현하는지 알려주기위해
    - autopct: 조각내에 표시될 비율의 문자열 형식. '%fmt문자' 
        - fmt문자: f(실수), d(정수), %% (%)
- 막대그래프를 비율로 보여준 것     `

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 10
b = 5.23
f'{a},{b}'
```

```python
'{},{}'.format(a,b)
```

```python
"%d-%.2f"  %(a,b)
```

```python
labels = ['사과','배','복숭아','수박','딸기']
amount=[20,10,40,5,60]


plt.figure(figsize=(10,10))
plt.pie(amount,labels = labels, autopct = '%.2f%%',
       explode = [0.2,0,0,0,0], shadow = True,
       textprops = {'fontsize':15})

plt.show()
```

# 5 히스토그램 그리기
## .1 히스토그램
- 표로 된 도수 분포표를 그래프로 나타낸 것.
    - 도수분포표: 특정 그룹안에 값의 빈도를 나타낸 표
    - 주로 연속형 자료를 특정 구간으로 나눠 그 빈도를 확인한다.
        - 빈도 확인이나 **분포**를 볼때 사용한다.
    - X축: 계급(변수구간)의 대표값, Y축: 빈도수
- `hist(data [, bins=계급개수)` 메소드를 사용
    - data: 리스트형의 객체를 전달한다.
- 연속형 데이터에 사용하는게 범주형.

# 6. 상자그래프(Boxplot) 그리기
## 6.1 상자그래프란
- 값들의 중간값, 사분위수, 최대/최소값을 보여주어 값들의 분포를 확인 할때 사용한다.
- boxplot(x, whis=1.5) 
    - x: 데이터셋
    - whis: 극단치 계산시 사용할 값(기본 1.5) " 정상범위를 좁게하고 싶을 때 whis값 작게때 whis값 작게
        - 극단치 : 1사분위와 3사분위에서 $IQR(3사분위수-1사분위수) * whis$ 범위 이상 떨어진 값들
- 범위적인걸로 보는것
- 이상치가 아닌 것으 ㅣ맥스값, 이상치가 아닌 것의 최소값을 보여줌 -> 분포에서 정상적인 값의 범위
- 이상치가 있는지도 확인 가능 -> 이상치는 대체하거나 결측치처리
    - 이상치때문에 정확한 결과를 얻을 수 없을 때 : 데이터를 더 많이 모으는 것도 방법(이상치가 정상치에 미치는 영향이 적어지니까)-> 하지만 이러지 못하는 경우도 있지

```python
data = np.random.randint(1,100,1000)
data[:5]=[200,250,-50,-100,100]
data[:10]
```

```python
plt.figure(figsize=(7,7))
plt.boxplot(data)

plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.boxplot(data, whis = 5,vert = False)
plt.show
```
