```python
# 최적화에서 경사하강법 사용시 알파라는 파라미터 조정해줘야(학습률학습률)
```

```python
def f(x):
    return (x-1)**2 + 2
```

```python
def fd(x):
    return 2*(x-1)
```

```python
import matplotlib as mpl
mpl.rcParams['font.family'] = 'malgum gothic'
```

```python
X = np.linspace(-3,4,100)

plt.figure(figsize = (10,10))
plt.plot(X, f(X), 'k-')

learning_rate = 1

w = 0
plt.plot(w,f(w),'go',markersize = 10)
plt.text(w+0.1,f(w)+0.1, '1차시도')
plt.plot(X, fd(w)+(X)+f(w), 'g--')
print('1차시도: w ={:.2f}, 기울기 = {:.2f}'. format(w,fd(w)))

w = w -learning_rate*fd(w)
plt.plot(w,f(w), 'ro', markersize = 10)
plt.text(w+0.1, f(w)+0.1, '2차시도')
plt.plot(X, fd(w)+(X-w)+f(w), 'r--')
print('2차시도:w ={:.2f}, 기울기 = {:.2f}'.format(w,fd(w)))

w = w - learning_rate*fd(w)
plt.plot(w, f(w), 'bo', markersize=10)
plt.text(w+0.1, f(w)+0.1, '3차시도')
plt.plot(X, fd(w)*(X-w) + f(w), 'b--')
print('3차시도: w={:.2f}, 기울기 = {:.2f}'.format(w, fd(w)))


w = w - learning_rate*fd(w)
plt.plot(w, f(w), 'mo', markersize=10)
plt.text(w+0.1, f(w)+0.1, '4차시도')
plt.plot(X, fd(w)*(X-w) + f(w), 'm--')
print('4차시도: w={:.2f}, 기울기 = {:.2f}'.format(w, fd(w)))


w = w - learning_rate*fd(w)
plt.plot(w, f(w), 'co', markersize=10)
plt.text(w+0.1, f(w)+0.1, '5차시도')
plt.plot(X, fd(w)*(X-w) + f(w), 'c--')
print('5차시도: w={:.2f}, 기울기 = {:.2f}'.format(w, fd(w)))

w = w - learning_rate*fd(w)
plt.plot(w, f(w), 'yo', markersize=10, alpha=0.5)
plt.text(w+0.1, f(w)+0.1, '6차시도')
plt.plot(X, fd(w)*(X-w) + f(w), 'y--')
print('6차시도: w={:.2f}, 기울기 = {:.2f}'.format(w, fd(w)))


plt.ylim(0,15)
plt.xlim(-4,10)
plt.xlabel('W')
plt.ylabel('Cost')
plt.grid(True)
plt.show()
```

```python
# 학습률(알파) 너무 작게주면 조정을 엄청 오래해도 적절 가중치 못찾아
#너무크면 밖으로 돌아서 못찾아.
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler # 얘도 선형방식을 사용해서 전처리해줘야 성능 향상
from sklearn.model_selection import train_test_split #데이터 나누고
```

```python
X,y = load_breast_cancer(return_X_y = True)
X_train, X_test, y_train, y_test = train_test_split(X,y, stratify = y, random_state = 1)
```

```python
#전처리
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#모델생성 및 학습
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

lr = LogisticRegression(random_state = 1)
lr.fit(X_train_scaled,y_train)
```

```python
pred_train = lr.predict(X_train_scaled)
pred_test = lr.predict(X_test_scaled)
```

```python
#평가
accuracy_score(y_train, pred_train), accuracy_score(y_test, pred_test)
```

# 군집 : 비지도학습. y(정답이 없는 것)
- 고객분류, 이상치탐지, 준지도학습에 사용
- k-mean : 가장 대표적 군집 알고리즘

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'margum gothic'
mpl.rcParams['axes.unicode_minus'] =True
```

```python
from sklearn.datasets import load_iris
data = load_iris()
X,y = data['data'], data['target']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 3,random_state = 1)

kmeans.fit(X_scaled)
```

```python
cluster = kmeans.labels_ # 학습한 데이터 셋 을 각각 어떤 그룹으로 나눴는지라벨 반환
cluster
```

```python
# 적절한 k값 찾기- inertia value
#k가 커질수록 이녀샤의 값은 작아져.체감적으로 작아짐.
# 이녀샤의 값이 작은 것이 군집화 잘 된것.
#KMean.inertia_ 속성으로 확인 가능
# 군집단위별로 이녀사 값이 급격히 떨어지는 지점이 적정 군집수라 판단가능
```

```python
kmeans.inertia_
```

```python
k_list = range(2,15)
inertia_list = []
for k in k_list:
    model = KMeans(n_clusters = k, random_state = 1)
    model.fit(X_scaled)
    inertia_list.append(model.inertia_)
```

```python
plt.figure(figsize = (10,20))

plt.plot(k_list, inertia_list)
plt.xticks(k_list)
plt.show()
```

```python
#평가 : 실루엣 지표
# 현재 속한 그룹의 이니셔(A)와 가장 가까운 그룹(B)의 이니셔를 비교해보자!
# 계수는 A-B/MAX(A,B) : 1D에 가까울수록 좋은 지표
#silhouette_samples( 개별 관측치의 실루엣 계수 반혼)
#silhouette_score(): 실루엣 계수들의 평균
```

```python
from sklearn.metrics import silhouette_score, silhouette_samples
```

```python
s_coef = silhouette_samples(X_scaled, cluster) #x값, 클러스터(븐류된 라벨들)
cluster
```
