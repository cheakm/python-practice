## 1.2 matplotlib 그래프 구성요소


- 아래 용어는 기억
- figure
    - 전체 그래프가 위치할 기본 틀
    - https://matplotlib.org/api/_as_gen/matplotlib.pyplot.figure.html
    - 위의 예시에서, figure는 하나의 도화지, 실제 그래프는 axes에다 그린 것.
       -> 하나의 공간에다 여러개 그래프를 그리기 위해서 나눈 것.(위에처럼 하나에 4개그래프 그릴때)
    - axes를 모은게 figure
- axes(subplot)
    - 하나의 그래프를 그리기 위한 공간
        - figure에 한개 이상의 axes(subplot)로 구성해서 각 axes에 그래프를 그린다.
    - https://matplotlib.org/api/axes_api.html
- axis 
    - 축 (x축, y축)
    - axis label (x, y) : 축의 레이블(설명) : x axis lable
- ticks : 축선의 표시 (이걸 grid로 보여줌)
    - Major tick
    - Minor tick
- title : 플롯 제목  -> axes의 이름
- legend (범례)
    - 하나의 axes내에 여러 그래프를 그린 경우 그것에 대한 설명

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize = (15,7),facecolor = 'gray')
axes1 = fig.add_subplot(1,2,1)
axes2 = fig.add_subplot(1,2,2)


axes1.plot([1,2,3,4,5], [10,20,30,40,50], label='line1') #plot: 선그래프
axes1.plot([1,2,3,4,5], [50,40,30,20,10], label='line2')
axes2.scatter(np.random.randint(100, size=50), np.random.randint(100, 200, size=50), color='r')

fig.suptitle('Example of Plot', size=25, color='blue') #size: 폰트크기, color: 글자색
axes1.set_title("PLOT 1", size=20)
axes2.set_title("Plot 2", size=20)

axes1.set_xlabel("X축", size=15) # 세부적인 설정 바꾸기
axes1.set_ylabel("Y축", size=15)
axes2.set_xlabel("가격1", size=15)
axes2.set_ylabel('가격2', size=15)

axes1.legend()
axes1.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4,5],
        [10,20,30,40,50])

plt.title('pyplot 예제', fontsize = 20)
plt.xlabel('x축값')
plt.ylabel('y축값')

plt.show()
```

- 하나의 figure에 여러 그래프 그리기

## 3.2. Figure 와 Axes 객체를 이용해 그리기

- Figure에 axes를 추가한 뒤 axes에 그래프를 그린다.
- axes 생성 방법
    - figure.add_subplot() 메소드 이용
        - figure를 먼저 생성후 axes 들을 추가
    - pyplot.subplots() 함수를 이용
        - figure와 axes배열을 동시에 생성

###  3.2.1 figure.add_subplot() 메소드 이용
- figure객체에 axes를 추가하는 형태
- nrows(총행수), ncols(총열수), index(axes위치) 지정

```python
fig = plt.figure(figsize = (10,10)) # figure 객체 생성

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

axes = [ax1,ax2,ax3,ax4]
for i, ax in enumerate(axes, start= 1):
     txt = f'ax{i}'
     ax.text(0.5,0.5, txt, fontsize = 25)
#for i in axes:
#    txt = i
#    ax.text(0.5,0.5, txt, fontsize = 25)
    
plt.show()
```

```python
fig = plt.figure(figsize = (10,7))

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,3)
ax3 = fig.add_subplot(1,2,2) # 크기를 자유롭게 지정 가능

ax1.scatter([1,2,3,4,5],[5,2,3,4,1])

for i ,ax in enumerate([ax1,ax2,ax3],start=1):
    ax.set_title(f'ax{i}')
    ax.text(0.5,0.5 ,f'ax{i}',fontsize = 20)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
                 
plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize = (10,10))

ax1= fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(3,3,3)
ax4= fig.add_subplot(3,3,6)
ax5 = fig.add_subplot(3,3,9)

for i,ax in enumerate([ax1,ax2,ax3,ax4,ax5],start=1):
    ax.set_title(f'ax{i}',fontsize = 20)
    ax.text(0.6,0.5, f'ax{i}', fontsize = 20)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    
plt.tight_layout()
plt.show()
```

### 3.2.2 pyplot.subplots()
- nrows, ncols 로 axes개수와 위치 지정
- 반환: figure와 axes(subplot)들을 담은 **ndarray**(1차원, 2차원배열 다 가능)
    - figure객체와 넘파이배열을 반환
- add_subplot이 배치를 좀 더 자유롭게 가능하지만 얘는 일률적으로 행과열을 동일한크기로만 분할 가능

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2,2, figsize = (10,10)) # 행,열, fig: figure객체 반환 보통 figure에 하는 것은 타이틀 붙여주는 정도
fig.suptitle('figure 제목',size = 20)
print(type(axes), axes.shape, type(axes[0,0]))

axes[0,0].scatter([1,2,3,4,5],[7,3,4,2,1])

axes[0,0].set_title('[0,0]')
```

## 4.1 색 지정
- color 또는 c 속성을 이용해 지정
- 색상이름으로 지정. 
    - 색이름 또는 약자로 지정 가능
    - 'red', 'r'
    
| 문자열 | 약자 |
|-|-|
| `blue` | `b` |
| `green` | `g` |
| `red` | `r` |
| `cyan` | `c` |
| `magenta` | `m` |
| `yellow` | `y` |
| `black` | `k` |
| `white` | `w` |


- HTML 컬러문자열
    - #으로 시작하며 RGB의 성분을 16진수로 표현
    - #RRGGBB 또는 #RRGGBBAA
    - #FF0000, #00FF00FA
- 0 ~ 1 사이 실수로 흰식과 검정색 사이의 회색조를 표시
    - 0: 검정, 1: 흰색

- https://matplotlib.org/examples/color/named_colors.html
- https://htmlcolorcodes.com/
    - picker, chart(코드), name(색이름) 제공사이트

- 0~f로 갈수록 농도가 진해지는 것. (rr gg bb 순서 마지막(AA) 나머지는 투명도, 0 이되면 그색깔 표현하지     않음
