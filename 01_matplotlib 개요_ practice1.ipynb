```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
plt.scatter(df.set_index('dataset').loc['I'].x, df.set_index('dataset').loc['I'].y)
plt.title('Dataset I')

plt.subplot(2,2,2)
plt.scatter(df.set_index('dataset').loc['II'].x, df.set_index('dataset').loc['II'].y)
plt.title('Dataset II')

plt.subplot(2,2,3)
plt.scatter(df.set_index('dataset').loc['III'].x, df.set_index('dataset').loc['III'].y)
plt.title('Dataset III')

plt.subplot(2,2,4)
plt.scatter(df.set_index('dataset').loc['IV'].x, df.set_index('dataset').loc['IV'].y)
plt.title('Dataset IV')

plt.show()
```

## 1.2 matplotlib 그래프 구성요소


- 아래 용어는 기억
- figure
    - 전체 그래프가 위치할 기본 틀
    - https://matplotlib.org/api/_as_gen/matplotlib.pyplot.figure.html
    - 위의 예시에서, figure는 하나의 도화지, 실제 그래프는 axes에다 그린 것.
       -> 하나의 공간에다 여러개 그래프를 그리기 위해서 나눈 것.(위에처럼 하나에 4개그래프 그릴때)
    - axes를 모은게 figure
- axes(subplot)
    - 하나의 그래프를 그리기 위한 공간
        - figure에 한개 이상의 axes(subplot)로 구성해서 각 axes에 그래프를 그린다.
    - https://matplotlib.org/api/axes_api.html
- axis 
    - 축 (x축, y축)
    - axis label (x, y) : 축의 레이블(설명) : x axis lable
- ticks : 축선의 표시 (이걸 grid로 보여줌)
    - Major tick
    - Minor tick
- title : 플롯 제목  -> axes의 이름
- legend (범례)
    - 하나의 axes내에 여러 그래프를 그린 경우 그것에 대한 설명

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize = (15,7),facecolor = 'gray')
axes1 = fig.add_subplot(1,2,1)
axes2 = fig.add_subplot(1,2,2)


axes1.plot([1,2,3,4,5], [10,20,30,40,50], label='line1') #plot: 선그래프
axes1.plot([1,2,3,4,5], [50,40,30,20,10], label='line2')
axes2.scatter(np.random.randint(100, size=50), np.random.randint(100, 200, size=50), color='r')

fig.suptitle('Example of Plot', size=25, color='blue') #size: 폰트크기, color: 글자색
axes1.set_title("PLOT 1", size=20)
axes2.set_title("Plot 2", size=20)

axes1.set_xlabel("X축", size=15) # 세부적인 설정 바꾸기
axes1.set_ylabel("Y축", size=15)
axes2.set_xlabel("가격1", size=15)
axes2.set_ylabel('가격2', size=15)

axes1.legend()
axes1.grid(True)
plt.show()
```

## 3.1 pyplot 모듈을 이용해 그리기
- pyplot 모듈이 그래프 그리는 함수와 Axes(Subplot) 설정 관련 함수를 제공

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4,5],
        [10,20,30,40,50])

plt.title('pyplot 예제', fontsize = 20)
plt.xlabel('x축값')
plt.ylabel('y축값')

plt.show()
```

- 하나의 figure에 여러 그래프 그리기
    - plt.subplot(row,columns,num) # 나 써브플롯(axes 만들꺼야.)

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (10,10))

plt.subplot(2,1,1)

plt.plot([1,2,3],[10,20,30])

plt.title('첫번째')

#서브플롯지정
plt.subplot(2,1,2)
plt.scatter([1,2,3],[10,20,30])
plt.show('두번째')
```
